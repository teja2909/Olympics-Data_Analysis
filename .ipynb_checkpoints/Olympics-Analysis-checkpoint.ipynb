```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("athlete_events.csv")
region_df = pd.read_csv("noc_regions.csv")
```

```python
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df.describe()
```

# Filtering for Summer Olympics:

```python
df = df[df['Season'] == 'Summer']
df.head()
```

```python
df.shape
```

```python
region_df.head()
```

# Merging Regional Data with NOC Codes

```python
df = df.merge(region_df, on='NOC', how='left')
df.head()
```

```python
df['region'].unique()
```

```python
df.isnull().sum()
```

# Removing Duplicate Entries

```python
df.duplicated().sum()
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.duplicated().sum()
```

# Encoding Medal Data Using One-Hot Encoding

```python
df['Medal'].value_counts()
```

```python
medal = pd.get_dummies(df['Medal']).astype(int)
medal
```

```python
df.shape
```

```python
df = pd.concat([df, medal], axis=1)
df.head()
```

```python
df.shape
```

# Medal Tally

```python
df.groupby('NOC').sum()[['Gold', 'Silver', 'Bronze']].sort_values('Gold', ascending=False).reset_index().head(25)
```

```python
df[(df['NOC'] == 'IND') & (df['Medal'] == 'Gold')]
```

### Extracting Unique Medal Tally Records

```python
medal_tally = df.drop_duplicates(subset=['Team', 'NOC', 'Games', 'Year', 'City', 'Sport', 'Event', 'Medal'])
```

```python
medal_tally = medal_tally.groupby('NOC').sum()[['Gold', 'Silver', 'Bronze']].sort_values('Gold', ascending=False).reset_index()
```

```python
medal_tally['Total'] = medal_tally['Gold'] + medal_tally['Silver'] + medal_tally['Bronze']
```

```python
medal_tally
```
